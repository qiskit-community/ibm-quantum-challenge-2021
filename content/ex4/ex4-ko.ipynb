ëª…ë ¹ë“¤ì€ ë°±ì—”ë“œì˜ native gate set, ì¦‰ ë‹¨ì¼ íë¹„íŠ¸ ê²Œì´íŠ¸ $X$ì˜ ì œê³±ê·¼ `sx`ì™€ $Z$ ì¶•ì— ëŒ€í•œ ë§¤ê°œ ë³€ìˆ˜í™”ëœ íšŒì „, `rz(Î¸)` ë° 2 íë¹„íŠ¸ controlled-NOT `cx`ìœ¼ë¡œ êµ¬ì„±ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ì—¬ê¸°ì—ëŠ” ì—¬ëŸ¬ ê°œì˜ ë‹¤ë¥¸ ê²Œì´íŠ¸ê°€ ë‚˜ì—´ë˜ì–´ ìˆìœ¼ë©°, íŠ¹íˆ ì˜¤ì¼ëŸ¬ ê°ì— ê·¼ê±°í•˜ê³  ê³§ ë”ì´ìƒ ì‚¬ìš©ë˜ì§€ ì•Šì„ ["U-gates"](https://qiskit.org/textbook/ch-states/single-qubit-gates.html#generalU3)ê°€ ì™„ì „ì„±ì„ ìœ„í•´ ì•„ë˜ì— ìš”ì•½ë˜ì—ˆìŠµë‹ˆë‹¤.

í•œ ê°€ì§€ ì£¼ì˜í•  ì ì€ ê° `sx`ëŠ” í•˜ë‚˜ì˜ ë¬¼ë¦¬ì  í„ìŠ¤ë¡œ êµ¬ì„±ë˜ë©° `rz`ëŠ” [frame changes](https://arxiv.org/abs/1612.00858)ì— ì†Œí”„íŠ¸ì›¨ì–´ì—ì„œ êµ¬í˜„ëœë‹¤ëŠ” ì ì…ë‹ˆë‹¤. `instruction_schedule_map`ì—ì„œ ì²˜ìŒë¶€í„° êµ¬ì¶•í•˜ëŠ” ëŒ€ì‹  ê¸°ë³¸ ì¸¡ì • í„ìŠ¤ë¥¼ ê²€ìƒ‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ëŠ” ë³´ì •ëœ í„ìŠ¤ë“¤ë¡œë¶€í„° ìœ ìš©í•œ ì‹œì‘ì ìœ¼ë¡œ ì‚¬ìš©í•  ìˆ˜ ìˆê³  ê·¸ ì´ìƒë„ ì¡°ì‘í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# ë°±ì—”ë“œë¡œë¶€í„° ë³´ì •ëœ ì¸¡ì • í„ìŠ¤ ê²€ìƒ‰
meas = inst_sched_map.get('measure', qubits=[qubit])
meas.exclude(channels=exc_chans).draw(time_range=[0,1000])
```

<div class="alert alert-block alert-success">

## 1ë‹¨ê³„ (íŠœí† ë¦¬ì–¼): $|0\rangle \rightarrow |1\rangle$ ì „ì´ ì°¾ê¸°

ì´ ë¶€ë¶„ì˜ ì½”ë“œëŠ” ìˆ˜ì •í•  í•„ìš”ê°€ ì—†ìŠµë‹ˆë‹¤. ëª¨ë“  ì…€ì„ ì‚´í´ë³´ê³  ë‹¨ê³„ë“¤ì„ í™•ì‹¤íˆ ì´í•´í•˜ì„¸ìš”. `spec01_scheds`ê°€ ì–´ë–»ê²Œ êµ¬ì¶•ë˜ëŠ”ì§€ ì£¼ì˜ ê¹Šê²Œ ì‚´í´ë´ì•¼ í•©ë‹ˆë‹¤. ë§ˆì§€ë§‰ ë¬¸ì œì—ì„œ ìœ ì‚¬í•œ í„ìŠ¤ ìŠ¤ì¼€ì¤„ì„ ì‘ì„±í•´ì•¼ í•©ë‹ˆë‹¤. í„ìŠ¤ ìŠ¤ì¼€ì¤„ì„ ì‘ì„±í•˜ëŠ” ë°©ë²•ì— ëŒ€í•œ ìì„¸í•œ ë‚´ìš©ì€ [ì´ íŠœí† ë¦¬ì–¼](https://qiskit.org/documentation/tutorials/circuits_advanced/06_building_pulse_schedules.html)ì—ì„œ ì°¸ì¡°í•˜ì‹¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    
    
</div>

```python
from qiskit.pulse import DriveChannel, Gaussian

# 01 ë° 12 specì— ëŒ€í•´ ë™ì¼í•œ spec í„ìŠ¤
drive_amp = 0.25
drive_duration = inst_sched_map.get('x', qubits=[qubit]).duration

# ë³´ì •ëœ ë°±ì—”ë“œ í„ìŠ¤ëŠ” ê³ ê¸‰ DRAG í„ìŠ¤ë¥¼ ì‚¬ìš©í•˜ì—¬ |2> ìƒíƒœë¡œì˜ ëˆ„ì¶œì„ ì¤„ì…ë‹ˆë‹¤.
# ì—¬ê¸°ì„œëŠ” ê°„ë‹¨í•œ ê°€ìš°ì‹œì•ˆ í„ìŠ¤ë¥¼ ì‚¬ìš©ë©ë‹ˆë‹¤

drive_sigma = drive_duration // 4 # DRAG í„ìŠ¤ëŠ” ì¼ë°˜ì ìœ¼ë¡œ 4*sigma ê¸¸ì´ì…ë‹ˆë‹¤. 
spec_pulse = Gaussian(duration=drive_duration, amp=drive_amp, 
                        sigma=drive_sigma, name=f"Spec drive amplitude = {drive_amp}")

# ì‹¤í—˜ì„ ìœ„í•œ ì£¼íŒŒìˆ˜ì˜ np ë°°ì—´ êµ¬ì¶•
spec_freqs_GHz = helper.get_spec01_freqs(center_frequency, qubit)

# ê¸°ë³¸ ìŠ¤ì¼€ì¤„ ë§Œë“¤ê¸°
# ë“œë¼ì´ë¸Œ ì±„ë„ì— ì‘ìš©í•˜ëŠ” ë“œë¼ì´ë¸Œ í„ìŠ¤ë¡œ ë¶€í„° ì‹œì‘í•©ë‹ˆë‹¤.
spec01_scheds = []
for freq in spec_freqs_GHz:
    with pulse.build(name="Spec Pulse at %.3f GHz" % freq) as spec01_sched:
        with pulse.align_sequential():
            # ë§ˆì§€ë§‰ì—ì„œ ë¬¸ì œë¥¼ í’€ê¸° ìœ„í•´ì„œëŠ” ì´ë¶€ë¶„ì— ì£¼ì˜í•´ì•¼ í•©ë‹ˆë‹¤.
            pulse.set_frequency(freq*GHz, DriveChannel(qubit))
            pulse.play(spec_pulse, DriveChannel(qubit))
            pulse.call(meas)
            
    spec01_scheds.append(spec01_sched)

# spec01 ìŠ¤ì¼€ì¤„ ê·¸ë¦¬ê¸°
spec01_scheds[-1].exclude(channels=exc_chans).draw(time_range=[0,1000])
```

```python
from qiskit.tools.monitor import job_monitor

# ì‹¤ì œ ë°±ì—”ë“œì—ì„œ ì‘ì—… ì‹¤í–‰
spec01_job = backend.run(spec01_scheds, job_name="Spec 01", **helper.job_params) 

print(spec01_job.job_id())
job_monitor(spec01_job)

# ëŒ€ê¸°ì—´ ì‹œê°„ì´ ë„ˆë¬´ ê¸´ ê²½ìš° ì‘ì—… idë¥¼ ì €ì¥í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# ì‘ì—…ì´ ì™„ë£Œë˜ë©´ ì‘ì—…ì„ ì–»ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# 'JOB_ID'ë¥¼ ë³¸ì¸ì˜ ì‘ì—… idë¡œ ë°”ê¾¸ì‹œê³  ì•„ë˜ ì½”ë“œë¥¼ ì£¼ì„ í•´ì œ í•´ì¤ë‹ˆë‹¤.
#spec01_job = backend.retrieve_job('JOB_ID')
```

### ë¶„ê´‘ê¸°(Spectroscopy) ë°ì´í„° ë§ì¶¤

```python
max_rabi_amp = 0.75
rabi_amps = helper.get_rabi_amps(max_rabi_amp)

rabi_scheds = []
for ridx, amp in enumerate(rabi_amps):
    with pulse.build(name="rabisched_%d_0" % ridx) as sched: # '0'ì€ ë¼ë¹„ì— í•´ë‹¹í•©ë‹ˆë‹¤
        with pulse.align_sequential():
            pulse.set_frequency(f01*GHz, DriveChannel(qubit))
            rabi_pulse = Gaussian(duration=drive_duration, amp=amp, \
                              sigma=drive_sigma, name=f"Rabi drive amplitude = {amp}")
            pulse.play(rabi_pulse, DriveChannel(qubit))
            pulse.call(meas)
    rabi_scheds.append(sched)

# ë¼ë¹„ ìŠ¤ì¼€ì¤„ ê·¸ë¦¬ê¸°
rabi_scheds[-1].exclude(channels=exc_chans).draw(time_range=[0,1000])
```
