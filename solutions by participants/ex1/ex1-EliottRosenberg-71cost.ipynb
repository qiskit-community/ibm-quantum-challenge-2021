I began with the decomposition of the Toffoli gate given here:
https://qiskit.org/textbook/ch-gates/more-circuit-identities.html

Then I substituted the decomposition of the Hadamard gate worked out earlier in this exercise. Then I combined two adjacent Rz rotations.

```python
from math import pi

qc = QuantumCircuit(3)

qc.rz(pi/2,2)
qc.sx(2)
qc.rz(pi/2,2)
qc.cx(1,2)
qc.rz(-pi/4,2)
qc.cx(0,2)
qc.rz(pi/4,2)
qc.cx(1,2)
qc.rz(-pi/4,2)
qc.cx(0,2)
qc.rz(pi/4,1)
qc.rz(3*pi/4,2)
qc.sx(2)
qc.rz(pi/2,2)
qc.cx(0,1)
qc.rz(pi/4,0)
qc.rz(-pi/4,1)
qc.cx(0,1)

qc.draw()
```

Notice that an Rz can be commuted past two CX gates if the rotation is on the target qubit. (If the control bit is 0, this is trivial; if the control bit is 1, this negates the angle twice.)
